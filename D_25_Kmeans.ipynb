```python
def randomm(df,k):
    klist = list(range(1,k+1))
    D = {}
    for a in klist:
        ls=[]
        for i in df.columns:  
            x=random.randint(df[i].min(),df[i].max())
            ls.append(x)
        D[a]= ls
    return  D
randomm(mallCust, 2)      
```

```python
class Clustering:
    def __init__(self,df,k):
        self.df=df
        self.k=k
    def randomm(self):
        klist = list(range(1,self.k+1))
        self.D = {}
        for a in klist:
            self.ls=[]
            for i in self.df.columns:  
                self.x=random.randint(self.df[i].min(),self.df[i].max())
                self.ls.append(self.x)
            self.D[a]= self.ls
        return  self.D 
       
```
